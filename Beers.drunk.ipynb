```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
file = 'drink-drinking.csv'
```

```python
df = pd.read_csv(file)
print(df.head())
```

```python
country_beer_consumption = df.groupby('country')[['beer_servings']].mean()
print(country_beer_consumption)
```

```python
by_country_median = df.groupby('country')[['beer_servings']].median()
print(by_country_median)
#not raw talk abt in the read me
```

```python
by_country_median.plot(kind = 'bar', title = 'amount of beers drunk in each country', ylabel = 'amount of beer drunk', xlabel = 'Country')
```

```python
x = df['country']
y = df['beer_servings']

plt.scatter(x, y)
plt.show()
```
